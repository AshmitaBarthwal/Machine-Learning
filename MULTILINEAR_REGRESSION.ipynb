```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices_2.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['area'],df['price'],color='red')
plt.show()
```

```python
plt.scatter(df['age'],df['price'],color='green')
plt.show()
```

```python
plt.scatter(df['bedrooms'],df['price'],color='orange') #phle independent aayega phir dependent variable aayega
plt.show()
```

```python
#independent variable denoted by x
x=df[['area','age','bedrooms']] #IV always atored as 2D
x
```

```python
#if there are many IV in bigger dataset then take only necessary columns as independent columns that will really affect DV
#dependent variable denoted by y
y=df['price']
y
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x,y)#it will calc coeff and intercept 
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x)
score=r2_score(y_pred,y)
score
```

```python
model.coef_#gives value of m1 m2 and m3
```

```python
model.intercept_
```

```python
model.predict([[3000,3,30]])#area,no of bedrooms,age
```

```python
area=int(input('Area :'))#dynamically values are given
broom=int(input('Bedrooms :'))
age=int(input('Age of the house'))
result=model.predict([[area,broom,age]])
print('the price of the house is :',result)
```
