```python
lr= linear_model.LinearRegression()#object pointing to linear regression method
```

```python
lr.fit(x,y) #fit method will calculate the value of m and b in Y=mx+b
```

```python
# finding value of m
lr.coef_
```

```python
lr.intercept_
```

```python
#calculating price manually
#price = m*area + b
#price = 135.79 *3300 + 180616.44
#price=628723.44
```

```python
lr.predict([[3300]]) #used 2 square brackets bcz of 2D passage of values
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#plt.scatter(df.year,df.per capita income (US$))
#plt.show()
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
```

```python
x= df[['year']] #two dimensional
x
```

```python
y = df['per capita income (US$)']
y
```

```python
lr= linear_model.LinearRegression()
```

```python
lr.fit(x,y) 
```

```python
lr.score(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.predict([[2020]])
```

```python
year=int(input('Enter the year in which you have to find the per capita income : '))
price=lr.predict([[year]])
#print('the per capita income is',per capita income (US$)'in year',year)
print(f'The per capita income is {price[0]} (US$) in year {year}.')
```
